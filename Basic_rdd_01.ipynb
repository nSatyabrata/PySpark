```
print("Count : ",rdd1.count())
print("First : ",rdd1.first())
print("Top : ",rdd1.top(2)) # top action : send data to driver -> order -> get top
```

```
# distinct
rdd1.distinct().collect()
```

```
# map
rdd_map = rdd1.map(lambda x: x**2)
rdd_map.collect()
```

```
# filter
rdd_odd_filter = rdd1.filter(lambda x: x%2)
rdd_odd_filter.collect()
```

```
# flatmap
rdd_flatmap = rdd1.flatMap(lambda x: (x+2,x*2))
print(rdd_flatmap.collect())
print("reduced sum : ",rdd_flatmap.reduce(lambda x,y: x+y))
```

```
# Descriptive stats
print(rdd1.max(),rdd1.min(),rdd1.mean(),rdd1.stdev(),rdd1.sum())
```

```
# mapPartitions
def partition_sum(partition):
    sum = 0
    for item in partition:
        sum = sum + item
    yield sum

rdd1.mapPartitions(partition_sum).collect()
```
